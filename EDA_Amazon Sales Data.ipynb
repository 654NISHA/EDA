# Amazon Sales Data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('amazon.csv')
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹',"").str.replace(",","")
```

```python
df['discounted_price'] = df['discounted_price'].astype(float)
```

```python
df['discounted_price']
```

```python
 df['actual_price'] = df['actual_price'].str.replace('₹',"").str.replace(',',"")
```

```python
df['actual_price'] = df['actual_price'].astype(float)
```

```python
df['actual_price']
```

```python
df['discount_percentage'] =df['discount_percentage'].str.replace("%","")
```

```python
df['discount_percentage'] = df['discount_percentage'].astype(int)
```

```python
df['discount_percentage']
```

```python
df['rating'] = df['rating'].replace(to_replace = {"|",np.nan})
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors = 'coerce')
```

```python
df['rating']
```

```python
df['rating_count'] = df['rating_count'].str.replace(',',"")
```

```python
df['rating_count'] = pd.to_numeric(df['rating_count'], errors = 'coerce')
```

```python
df['rating_count']
```

```python
df.info()
```

### Que3.What is the distribution of discounted prices vs. actual prices?
### Insigths: The distribution of discounted prices versus actual prices indicates a positive correlation between the two. This suggests that as the actual price of a product increases, the corresponding discounted price tends to increase as well. This positive correlation implies that higher-priced items tend to have higher discounts applied to them, maintaining a consistent relationship between actual and discounted prices across the dataset.

```python
df['actual_price'].corr(df['discounted_price'])
```

```python
plt.scatter(df['actual_price'], df['discounted_price'])
plt.title('Disribution of discounted price vs actual price')
plt.xlabel('Actual price')
plt.ylabel('Discounted price')
plt.show()
```

### Que4.  How does the average discount percentage vary across categories?

```python
average_per_dis = df.groupby(['category'])['discount_percentage'].mean()
```

```python
plt.figure(figsize=(10, 6))
average_per_dis.plot(kind='bar')
plt.title('Average Discount Percentage Across Categories')
plt.xlabel('Category')
plt.ylabel('Average Discount Percentage')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
